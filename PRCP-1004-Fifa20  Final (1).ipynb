- The word **forward** is used both as a general term for attacking players and a specific position.

- **Center forwards** are positioned right behind strikers. They are expected to receive passes from their teammates and score goals or create assists for others. In addition to the skills expected of strikers, center forwards also need to be good at passing.

- **Right forwards and left forwards** are positioned to the right and left of center forwards. They have the same expectations as center forwards.

- **Wingers** are positioned near the touchlines. Their main responsibilities are to create chances for strikers and forwards by breaking through and crossing from the right and left sides of the field, and to score goals themselves. Wingers need to have good dribbling, acceleration, passing, and crossing skills.

### Handling Missing Values

From the table, we can see that J.Oblak and M. ter Stegen have a missing value for the first six attributes whereas the remaining players have a missing value for the last six attributes. J. Oblak and M.ter Stegen are goalkeepers, meaning that there is no information about their pace, dribbling or shooting since goalkeepers do not perform such skills. Likewise, other players such as L.Messi do not have gk_diving or gk_positioning attributes as such skills are not required for him. Since all the ratings are out of 100, we are replacing such NA with 1 to indicate that the particular players have a very low level of skills for such attributes.

```python
# Replace 'NA' values with 1
k_fifa.fillna(1, inplace=True)
```

```python
# Replace non-numeric values (e.g., strings) with NaN
k_fifa = k_fifa.apply(pd.to_numeric, errors='coerce')

# Replace NaN values with 1
k_fifa.fillna(1, inplace=True)
```

# 

### k-means clustering

We will assign three cluster for three most basic position of football (Attack, defense and goalkeeper).

```python
# Perform k-means clustering with k=3
kmeans = KMeans(n_clusters=3, random_state=100)
clustered = kmeans.fit_predict(k_fifa)
```

```python
# Count the number of players in each cluster
cluster_counts = pd.Series(clustered).value_counts()
```

```python
# Display cluster counts
print(cluster_counts)
```

# 

### Cluster results

```python
# Add the cluster labels to the original dataset
fifa20['Cluster'] = clustered
```

```python
# some players in each cluster 

# Create a dictionary to store players by cluster
clustered_players = {}
for cluster_id in range(3):
    cluster_data = fifa20[fifa20['Cluster'] == cluster_id].head(10)[['short_name', 'Cluster']].values.tolist()
    clustered_players[f'Cluster {cluster_id}'] = [row[0] for row in cluster_data]

# Display the table using tabulate
table = tabulate(clustered_players, headers=['Cluster 1', 'Cluster 2', 'Cluster 3'], tablefmt='pretty')

# Print the table
print(table)
```

### Sub Task 4.2

**Ideal budget to create a competitive team (at the level of top n teams in Europe) and at which point the budget does not allow to buy significantly better players for the 11-men lineup. An extra is the same comparison with the Potential attribute for the lineup instead of the Overall attribute**

```python
# Group the data by the 'club' column and sum the 'value_eur' for each club
club_budgets = data.groupby('club')['value_eur'].sum()

# Find the top 10 teams with the highest budgets
top_teams = club_budgets.nlargest(10).reset_index()

# Create a list of lists for tabulate
table_data = []
for index, row in top_teams.iterrows():
    table_data.append([index + 1, row['club'], f'{row["value_eur"]:.2f} EUR'])

# Display the top 10 teams and their budgets in tabulate format with an index number column
print(tabulate(table_data, headers=['Index', 'Team', 'Budget'], tablefmt='grid'))
```

```python
# Sort players by overall rating in descending order and select the top 11
top_players_by_overall = data.sort_values(by='overall', ascending=False).head(11)

# Sort players by potential rating in descending order and select the top 11
top_players_by_potential = data.sort_values(by='potential', ascending=False).head(11)

# Calculate the budget required for the best team based on overall rating (only value_eur)
budget_by_overall = top_players_by_overall['value_eur'].sum()

# Calculate the budget required for the best team based on potential rating (only value_eur)
budget_by_potential = top_players_by_potential['value_eur'].sum()

# Create a table for tabulate
table = [["", "Approximate Ideal Budget for Competitive Lineup"],
         ["Overall:", f"{budget_by_overall:.2f} EUR"],
         ["Potential:", f"{budget_by_potential:.2f} EUR"]]

# Print the results using tabulate
print(tabulate(table, headers="firstrow", tablefmt='grid'))
```

```python
# Sort teams that can afford the budget by budget in descending order
top_teams_by_overall_budget = top_teams[top_teams['value_eur'] >= budget_by_overall].sort_values(by='value_eur', ascending=False)
top_teams_by_potential_budget = top_teams[top_teams['value_eur'] >= budget_by_potential].sort_values(by='value_eur', ascending=False)

# Create tables for tabulate
table_overall_budget = top_teams_by_overall_budget.reset_index().rename(columns={'index': 'Index', 'club': 'Team', 'value_eur': 'Budget (Overall)'})
table_potential_budget = top_teams_by_potential_budget.reset_index().rename(columns={'index': 'Index', 'club': 'Team', 'value_eur': 'Budget (Potential)'})

# Print the results using tabulate
print("\nTeams that can afford the Budget for Competitive Lineup (Overall = ",budget_by_overall,"):")
print(tabulate(table_overall_budget, headers='keys', tablefmt='grid'))

print("\nTeams that can afford the Budget for Competitive Lineup (Potential = ",budget_by_potential,"):")
print(tabulate(table_potential_budget, headers='keys', tablefmt='grid'))
```
